```python
Title:  'Homework 2: Jupyter Notebook and Introduction to Statistical Learning'
Subtitle: 'ISTA 421 -- Fall 2024'
Author: "Celso Haros | College of Information"

#Question 1
#For each of four scenarios listed below, indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.

#The sample size (number of observations; n) is extremely large, and the number of predictors (features; p) is small.

#ANSWER: The inflexible method would fit here. Since there is more data and less features in this substrate, 
#the inflexible method would garner lower variance and better generalize the data, instead of overfitting it as the flexible method would. 

#The number of predictors (p) is extremely large, and the number of observations (n) is small.

#ANSWER: Given the extreme number of predictors here, the flexible method would be worse as it would overfit the predictors due to the small amount of data (n). 
#This would lead to high variance and to poor performance since data is too sparse and there is much too little to decide on given "p".

#The relationship between the predictors and response is highly non-linear.''

#ANSWER: The flexible method is better as when it comes to a highly non-linear relationship, the data becomes complex.
#It offers better modeling over interactions and unforeseen variations between both predictors and responses. It stays true to its name as "Flexible". 

#The variance of the error terms (i.e.) is extremely high.

#ANSWER: The inflexible method is better, as we are dealing with high variance of error terms (synonymous to noisy data).
#The flexible method would overfit the noisy data, whereas inflexible methods are less sensitive to this noisy data and create far less fluctuations throughout. It's just simpler.

#Question 2
#In a few sentences, please answer the following questions to the best of your knowledge. Feel free to conduct additional research and cite your sources.

#Briefly explain the curse of dimensionality, provide a hypothetical example illustrating the concept, and list at least one potential way that is generally used to handle it when using machine learning models.

#ANSWER: The curse of dimensionality refers to the phenomenon in which the volume of feature space increase exponentially in proportion to the number of features (or better known as predictors). 
#Data is spread out due to an abrupt increased number of possible combinations and patterns are difficult to find, in result, insights based on the data become less likely. This causes issues with a model's ability to generalize efficiently. 

#Explain why the relationship between model error and complexity differs when patterns are examined using training and test datasets.

#ANSWER: The relationship between both model error and complexity differs in training and test datasets due to bias-variance tradeoff which asserts that when a model's complexity increases, bias inversely decreases, and variance increases. 
#Like with a u-shaped curve, and how it's left side may depict a lower complexity model entailing underfitting, high bias, and low variance. 
#The right side would thus entail the opposite; a highly complex model with high variance and low bias that leads to overfitting. 

#To the best of your knowledge, distinguish between training, test, and validation datasets. Briefly describe the importance of each in the context of Machine Learning.

#ANSWER: A training dataset trains a machine learning model to understand patterns and relationships granted from the training data itself. 
#A validation dataset separates datasets during model training in order to fine tune the model's parameters and gauge its performance at its optimum state.
#A test dataset serves as a means to evaluate a model's performance post-training. 

#Briefly discuss the consequences of overfitting and underfitting.

#ANSWER: Overfitting will lead to too much of a complex model that becomes extremely noisy, leading to poor generalization with unseen data and ergo a more probable chance of reaching a high-test error.
#Underfitting on the other hand occurs when a model is too simple to understand underlying patterns or motives within a dataset. 
#The consequence of this is primarily inaccuracy caused by a blatant lack of understanding of the data.

#Question 3
#Explain whether each of the scenarios presented below is a classification or regression problem. Indicate whether the situation is mostly interested in conducting inference (explaining patterns) or prediction. Finally, indicate the number of observations (n) and features (p) associated with each of the .

#We collect a set of data on the top 500 firms in the US. For each firm we record profit, number of employees, industry and the CEO salary. We are interested in understanding which factors affect CEO salary.

#ANSWER: This is a regression problem as its response variable (CEO's salary) is quantitative. We are primarily attempting to formulate inference in relation to what factors may affect the response variable. 
#We know (n) to be 500, and (p) to be 3 (profit, employee count, industry). 

#We are considering launching a new product and wish to know whether it will be a success or a failure. We collect data on 20 similar products that were previously launched by different companies. For each product we have recorded its type, whether it was a success or failure, price charged for the product, marketing budget, competition price, and ten other variables.

#ANSWER: This is a classification problem since its response failure is dual, success or failure; ergo categorical.
# We are looking to predict the outcome of the incoming product based on previous trial and error, and we know (n) to be 20, and (p) to be 13 (product, marketing budget, competition price, and ten other variables). 

#We are interested in predicting the % change in the USD/Euro exchange rate in relation to the weekly changes in the world stock markets. Hence we collect weekly data for a given year. For each week we record the % change in the USD/Euro, the % change in the US market, the % change in the British market, and the % change in the German market.

#ANSWER: This is a prime example of regression, as its response variable is quantitative referring to percentage change. 
#We are looking to predict the percentage change in USD/Euro market fluctuations, and we know that (n) is 52 for every week of the year and that (p) is 3 in correlation to the given number of markets. 

#Question 4
#Let's now revisit the bias-variance decomposition.

#Provide a sketch of typical (squared) bias, variance, training error, test error, and irreducible (sometimes called Bayes error) error curves, on a single plot, as we go from less flexible statistical learning methods towards more flexible approaches. The x-axis should represent the amount of flexibility in the method, and the y-axis should represent the values for each curve. There should be five main curves. Make sure to label each one. Now, add two arrows (parallel to the X axis) indicating the direction of increase in over- and under-fitting, respectively. Finally, label the point, also in the X axis, where model complexity is optimal.
#NOTE: Please either draw your sketch on a piece of paper and then scan, or draw it with matplotlib or seaborn. If you draw by hand and then scan, you'll need to insert the image using the relevant code AND submit the image file along with your homework.

#ANSWER:
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# X Values representing flexibility from low to high
x = np.linspace(1, 10, 100)

# General curves based on bias-variance tradeoff
squared_bias = 1 / (x ** 0.5)  # Squared bias decrease based on flexibility
variance = (x ** 1.5) / 20  # Variance increase based on flexibility
irreducible_error = np.ones_like(x) * 0.5  # Constant irreducible error/Bayes error
training_error = (1 / x)  # Training error decreases based on flexibility
test_error = squared_bias + variance + irreducible_error  # Test error

# Other curves
plt.figure(figsize=(10, 6))
plt.plot(x, squared_bias, label="Squared Bias", color='blue')
plt.plot(x, variance, label="Variance", color='green')
plt.plot(x, test_error, label="Test Error", color='red', linestyle='--')
plt.plot(x, training_error, label="Training Error", color='purple', linestyle=':')
plt.plot(x, irreducible_error, label="Irreducible Error", color='orange', linestyle='-.')

# Optimal Complexity Point
optimal_complexity_x = 4.5
plt.axvline(x=optimal_complexity_x, color='gray', linestyle='--', label="Optimal Complexity")

# Arrows where overfitting and underfitting
plt.annotate('Underfitting', xy=(1.5, 1.4), xytext=(3, 2),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Overfitting', xy=(8, 4), xytext=(6, 5),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Add another arrow for overfitting
plt.annotate('Overfitting', xy=(8, 2), xytext=(7, 2.2),
             arrowprops=dict(facecolor='black', shrink=0.05))

# For Title and Labels (Key)
plt.xlabel("Model Flexibility")
plt.ylabel("Error")
plt.title("Bias-Variance Decomposition")
plt.legend()
plt.show()

# Explain what the conceptual importance of acknowledging the existence of irreducible error is in the context of Machine Learning.

#ANSWER: Irreducible error is the noise within the data that can't be reduced regardless of how complex the model may be.
# It's a reality check for the incapability of even machines to avoid a margin of error. 

# Finally, briefly explain why each of the five curves has the shape displayed.

#ANSWER: 
# The blue curve represents the squared bias, it decreases inversely to model flexibility because complex models usually better fit training data and serve best when the intent is to capture patterns. 
# The green curve is the variance curve, which increases alongside model flexibility as complex models become extremely sensitive to perpetual influx and activity of training data. When this data grows, overfitting occurs. 
# The test error curve is characterized by a red dashed curve, which follows a U-shape reminiscent of the aforementioned tradeoff between bias and variance. 
# The purple dotted curve is the training error curve which is characterized by a proportionate decrease with increased flexibility. This is because a complex model will best fit the training data due to specificity. 
# The orange dot-dash curve represents Irreducible Error, and remains constant as the range of flexibility is presented as noise, this as mentioned earlier can't be reduced regardless of the model. It serves to gauge just how well a model may perform.

#Question 5
#You will now think of some real-life applications for statistical learning.

#Describe three real-life applications in which classification might be useful. Describe the response variable, as well as the potential predictors. Is the goal of each application inference or prediction? Explain your answer.

#ANSWER: 
# 1. Example 1: Identifying Wildlife Specifies in a forest. The response variable is the type of species of an animal, the predictors are size, sound, and habitat, and the goal is to predict (prediction) because we classify the documented species based on their environment and prior data. 
# 2. Example 2: When Will Loan Default? Response Variable is whether or not it is default (which is categorical), predictors are credit history, income, and the loan balance, and the goal is to predict what applicants may be identified as "high-risk". 
# 3. Example 3: What is Fake News? The response variable is categorical, i.e. fake or not fake. The predictors are credibility of sources, word choice, and word count. The goal is to predict whether or not a news article is authentic or not. 

#Describe three real-life applications in which regression might be useful. Describe the response, as well as the predictors. Is the goal of each application inference or prediction? Explain your answer.

#ANSWER:
# Example 1: Predicting a movie's box office revenue which's RV (Response Variable from now on) is box office revenue, its predictors are cast, genre, movie length, and marketing budget, and its goal is to predict a movie's future success. 
# Example 2: Fitness: The response variable here can vary, but let's choose achieving a personal record (PR) in the bench press. The predictor would be diet, recovery, and exercise programming. The goal is to predict what amount PR should be attempted at the end of the period. 
# Example 3: Climate Change's Effect on Nature: The RV here is how much crop fields yields. The predictors are time, temperature, and land quality. The goal is inference, ergo to understand the effect from climate change onto crops. 

#Question 6
#Lets now review some ethical considerations associated with the development and implementation of Machine Learning algorithms. Please answer each question with as much detail as possible (one or two short paragraphs per question should suffice). Feel free to conduct additional research if you think its necessary. Im looking for well-supported arguments. Your grade wont be based on whether or not I agree with your position. Instead, do your best to provide a thoughtful and clear answer.

#From your perspective, describe how unbiased must an algorithm be before it can be deployed in the real world? Is that level commonly achieved, discussed, or examined?

#ANSWER: Algorithms must be unbiased in order to prevent harm. Attaining fairness is difficult when it comes to biased data, which by 2024 has begun to an increasingly held debate amongst those who use, regulate, or manage AI. 
# However, complete removal of bias is extremely difficult to attain, as in much of data, bias is inherent in more ways than one, although, in the real world, they should not discriminate against sex, identity, or race; that's the minimum. 

#Research labs and companies generally invest the most in improving and developing ML algorithms. Do you think that society in general should also have immediate and equal access to these developments and to their benefits? How would you balance these two goals (e.g. profit and well-being)? Similarly, do you think ML already affects (or will affect) inequalities?

#ANSWER: I believe that society MUST have broader access to ML developments, regardless of what balance is found between either profit, yet placing societal consideration under a larger pilonce. It becomes pivotal, as AI and ML is the future of technology and clearly the next step human technology must strive to reach. 
# Innovation, reduction of biases, automation, the only struggle here becomes the capital required to fund it all, but it must be seen as a necessity by now. 

#What do you see as a solution for problems associated with increasing automation and efficiency, both currently and in the future?

#ANSWER: A probable solution would be through the very issue at hand; automation; upskilling and retraining today's workforce based on a perpetually augmenting model is imperative when it comes to mitigating what automation may do other jobs.
# I firmly believe that ML is for the people, used to augment, not replace. 

#We discussed different trade-offs between models in class  one of these was related to model interpretability. Why do you think it is important for machine learning models to be interpretable? Provide at least two reasons. Among the reasons that you listed, which one do you believe is the most important? Explain.

#ANSWER: Transparency and accountability are key for model interpretability; trust is a key virtue when it comes to improving models. 
# This is founded through proper distinction and definition of the processes at hand, especially to those less aware of jargon relevant to the field like stakeholders, who hold key decisions over the very constructs and financial processes ML may automate for them in service. 

#Now, between model performance and model interpretability: which of these two qualities do you think is more important. Explain.

#ANSWER: Interpretability becomes critical in industries such as that of health care; where data is not only means to generating or quantifying a company's efficiency in a business context, but also in treating the people, ethical considerations, and saving lives. 
# Model performance is pivotal towards generating adept predictions, emphasizing its accuracy, but something is better than nothing; ergo I side with interpretability, or if all else fails, the same ability to teach others to become literate in interpretability of fundamental or relevant to their work-force ML systems.

#Particular examples in history show that different algorithms and datasets were originally developed/simulated/compiled with goals related to increasing systemic inequalities. For instance, lets take a quick look at the iris dataset (Fisher 1936). First, how many hits do you get after searching for iris dataset tutorial in Google (feel free to try any similar or more systematic search queries)? Based on this search, list at least two different modern uses of the dataset.

#ANSWER: After conducting specific research on the Iris Dataset Tutorial, I managed to discover thousands of hits. 
# A modern use of this is found in creating a foundational metric one can use to generally compare two algorithms throughout the context of various academic paper. Another example is teaching fundamental ML classification techniques. 

# Applied Programming Questions
# Question 7
# This exercise relates to the College data set, which can be found in the file College.csv. It contains a number of variables for 777 different universities and colleges in the US. The variables are:

#Private : Public/private indicator
#Apps : Number of applications received
#Accept : Number of applicants accepted
#Enroll : Number of new students enrolled
#Top10perc : New students from top 10 % of high school class
#Top25perc : New students from top 25 % of high school class
#F.Undergrad : Number of full-time undergraduates
#P.Undergrad : Number of part-time undergraduates
#Outstate : Out-of-state tuition
#Room.Board : Room and board costs
#Books : Estimated book costs
#Personal : Estimated personal spending
#PhD : Percent of faculty with Ph.D.s
#Terminal : Percent of faculty with terminal degree
#S.F.Ratio : Student/faculty ratio
#perc.alumni : Percent of alumni who donate
#Expend : Instructional expenditure per student
#Grad.Rate : Graduation rate
#Use the Pandas read_csv() function to read the data into Python. The dataset can be found on the ISLP website, which you downloaded in homework 1. Read dataset into Python and save it as college.

#Note that the first column is just the name of each university. Although we are not going to use this column, it may be handy to have these names for later. First, assign the first column in college to the index of the dataframe. You can use the set_index() function in Pandas. . Second, delete the first column of college and overwrite the same object (college).
#Now you should see that the first data column in college is Private. Note that another column labeled Unnamed now appears before the Private column. However, this is not a data column but rather the name that Python is giving to each row. Use the describe() function to produce a numerical summary of the variables in the data set.
#Use the pandas.plotting.scatter_matrix() function to produce a scatterplot matrix of the first ten columns or variables of the data. You can reference the first ten columns of a dataframe using college.iloc[:, :10].
#NOTE: iloc is for index and loc is for value. Be sure you've imported matplotlib.
#Use the boxplot() function to produce side-by-side boxplots of Outstate versus Private.
#Create a new binary variable called Elite, by binning the Top10perc variable to college. We are going to divide universities into two groups based on whether or not the proportion of students coming from the top 10% of their high school classes exceeds 50%.
#Use the describe() or value_counts() function to see how many elite universities there are. Now use the boxplot() function to produce side-by-side boxplots of Outstate versus Elite.
#Use the hist() function to produce some histograms with diering numbers of bins for a few of the quantitative variables. You may find the plt.subplot() function useful to create multiple plots in one figure. Create such a figure with four subplots.

#ANSWER:
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

file_path = r'C:\Users\haros\Downloads\College.csv'
college = pd.read_csv(file_path)

college.set_index(college.columns[0], inplace=True)

if 'Private' not in college.columns:
    print("Error: 'Private' column not found post-indexing.")
else:
    print("'Private' column found")

# Summary
print(college.describe())

# Scatterplot matrix first ten columns
scatter_matrix(college.iloc[:, :10], figsize=(10, 10))
plt.show()

# Boxplot of Outstate and Private
college.boxplot(column='Outstate', by='Private', grid=False)
plt.title('Outstate Tuition by Private vs Public')
plt.suptitle('')
plt.show()

# New binary 'Elite'
college['Elite'] = (college['Top10perc'] > 50).astype('category')

# Boxplot of Outstate and Elite
college.boxplot(column='Outstate', by='Elite', grid=False)
plt.title('Outstate Tuition by Elite vs Non-Elite')
plt.suptitle('')
plt.show()

# Histograms for quantitative variables
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
college['Apps'].hist(bins=20, ax=axes[0, 0])
axes[0, 0].set_title('Applications Received')
college['Accept'].hist(bins=20, ax=axes[0, 1])
axes[0, 1].set_title('Applicants Accepted')
college['Enroll'].hist(bins=20, ax=axes[1, 0])
axes[1, 0].set_title('Students Enrolled')
college['Outstate'].hist(bins=20, ax=axes[1, 1])
axes[1, 1].set_title('Out-of-State Tuition')
plt.tight_layout()
plt.show()

# Summary Elite universities
print(college['Elite'].value_counts())

#Continue exploring the data, and provide a brief summary of what you discover. Some quick ideas: (1) explore graduation rates and briefly talk about what might be driving differences between universities. (2) Compare instructional expenditure per student and graduation rates.

#ANSWER: The average rate of graduation is an unprecedented 65.46% with a standard deviation of 17.17%, though there are anomalies with extreme lows and extreme, unrealistic highs surpassing even the 100% mark. 
# The scatter plot depicts instructional expenditure as a positive relationship between it and the graduation rate of the school. We can also see through the box plot that private universities have higher tuition (outstate) than public ones, which may entail better funding for the prior. 

#Are there any other features, not taken into account in the dataset, that you think might be important for understanding whether a school is classified as elite or not?

#ANSWER: The summary statistics depict observed instructional expenditure per student on average is $9660, reaching an extreme maximum of $56233. It's already been established that there is a positive relationship between both heightened instructional expenditure and graduation rates.
# We do see a trend through the scatter plots correlating the same higher instructional expenditure to higher graduation rates. How many students ranking in the top 25th percentile to top 10 percentile in their class attending a specific school may also attribute to this. So, graduation rates, quality of students, and funding/resources.

#Question 8
#This exercise involves the auto data.

#Make sure that the missing values have been removed from the data. Which of the predictors are quantitative, and which are qualitative?

import pandas as pd

auto = pd.read_csv(r'C:\Users\haros\Downloads\Auto.data', delimiter=r"\s+", na_values='?')

auto = auto.dropna()

quantitative = auto.select_dtypes(include=['float64', 'int64']).columns
qualitative = auto.select_dtypes(include=['object']).columns

print("Quantitative Predictors:", quantitative)
print("Qualitative Predictors:", qualitative)

range_quantitative = auto[quantitative].agg(['min', 'max'])
print("Range of each quantitative predictor:\n", range_quantitative)

mean_std_quantitative = auto[quantitative].agg(['mean', 'std'])
print("Mean and Standard Deviation of each quantitative predictor:\n", mean_std_quantitative)

auto2 = auto.drop(auto.index[9:85])

range_auto2 = auto2[quantitative].agg(['min', 'max'])
mean_std_auto2 = auto2[quantitative].agg(['mean', 'std'])

print("Range of each quantitative predictor in auto2:\n", range_auto2)
print("Mean and Standard Deviation of each quantitative predictor in auto2:\n", mean_std_auto2)

import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(auto[quantitative])
plt.show()



#ANSWER: 

#What is the range of each quantitative predictor? You can answer this using the min() max() describe() apply() agg() or other functions.

#ANSWER:
#mpg: 9.0 to 46.6
#cylinders: 3 to 8
#displacement: 68.0 to 455.0
#horsepower: 46.0 to 230.0
#weight: 1613 to 5140
#acceleration: 8.0 to 24.8
#year: 70 to 82
#origin: 1 to 3

#What is the mean and standard deviation of each quantitative predictor?

#ANSWER: 
mpg:
Mean: 23.45
Std: 7.81

cylinders:
Mean: 5.47
Std: 1.71

displacement:
Mean: 193.53
Std: 104.27

horsepower:
Mean: 104.47
Std: 38.49

weight:
Mean: 2977.58
Std: 849.40

acceleration:
Mean: 15.54
Std: 2.75

year:
Mean: 75.99
Std: 3.69

origin:
Mean: 1.57
Std: 0.80

#Now remove the 10th through 85th observations from auto. Save this as a new object called auto2. What is the range, mean, and standard deviation of each predictor in auto2?

#ANSWER: 
mpg:
Range: 11.0 - 46.6
Mean: 24.41
Std: 7.87

cylinders:
Range: 3 - 8
Mean: 5.37
Std: 1.59

displacement:
Range: 68.0 - 455.0
Mean: 187.71
Std: 99.89

horsepower:
Range: 46.0 - 230.0
Mean: 101.73
Std: 37.84

weight:
Range: 1649 - 4997
Mean: 2935.97
Std: 811.30

acceleration:
Range: 8.0 - 24.8
Mean: 15.71
Std: 2.70

year:
Range: 70 - 82
Mean: 76.26
Std: 3.68

origin:
Range: 1 - 3
Mean: 1.61
Std: 0.82

#Using the full data set (auto), investigate the predictors graphically, using scatter plots or other tools of your choice. Create some plots highlighting the relationships among the predictors. How could they be used for linear regression models?

#ANSWER:
# By using a scatter plot matrix, quantitative predictors such as mpg is negatively correlated with several predictive values are weight, displacement, and horsepower. A positive correlating concerning mpg can be seen with acceleration. 
# The negatively correlated relationships mentioned can be useful to a linear regression model due to their nature as, once more, predictors. 

#Suppose that we were interested in predicting gas mileage (mpg) on the basis of the other variables. Do your plots suggest that any of the other variables might be useful in predicting mpg? Justify your answer.

#ANSWER: 
# Yes, the scatter plots do suggest that variables such as weight, displacement, and horsepower are useful in predicting mpg as they serve as features with visible correlation, albeit negative.
# As any of the three values increase, mpg decreases. Acceleration has a positive correlation with mpg entailing the opposite. 

#Your Questions
#Use your scatterplot matrix visuals to ask three questions, 1 question per type:

#Inquisitive (curious/ clarifying): \

#QUESTION: Does the relationship between mpg and horsepower remain static throughout the years, or are there any changes to take notice of? 

#Critical (Verifying, proving, critique):\

#QUESTION: Given the negative correlation between mpg and weight, is it possible that other variables further influence this negative relationship between the two? Is multicollinearity part of that truth? 

#Applicative: \

#QUESTION: How can a car manufacturer best put into practice data gathered regarding the relationship between mpg, horsepower, and weight, to optimally design vehicles that seek to comply with modern fuel efficiency regulations whilst prioritizing performance? 

#Extra credit (2 points)
#What are the most highly-correlated variables in the College dataset? Justify your answer.
#The most correlated values in the College dataset are the Top10perc and Top25perc variables as their values overlap between one another, with the 10% logically and numerically being a subset of the top 25%. 
#If one goes up, so does the other. 




```
